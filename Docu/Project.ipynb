# 项目软件需求文档

## **项目 1：拼车应用（Car Sharing Application）**

### **1. 系统概述**
本项目为**拼车服务平台**，用户可以预约、提供和管理拼车，系统管理员负责管理用户和订单。

### **2. 主要功能**
#### **用户功能**
- **用户注册 & 认证**
  - 用户需要注册并登录才能使用系统。
  - 采用安全的密码存储方式（哈希加密）。

- **拼车预订系统**
  - 用户可以根据位置、日期、时间、乘客数量搜索拼车。
  - 用户可以预订和取消行程。
  - 允许用户设置拼车偏好（如行李空间、无障碍需求）。

- **行程管理**
  - 用户可以发布拼车信息，设定起点、终点、时间和可载人数。
  - 用户可以更新行程信息。

- **沟通系统**
  - 用户可以相互发送消息，讨论拼车细节。
  - 提供行程确认和取消的通知提醒。

- **费用计算与折扣**
  - 根据距离计算费用。
  - 提供常规折扣（如每周 4 次通勤可享受折扣）。

- **行程记录 & 统计**
  - 用户可以查看历史行程、总费用和总出行距离。

- **评价与问题反馈**
  - 用户可以对行程评分，并报告相关问题。

#### **管理员功能**
- **用户和行程管理**
  - 查看所有行程（可用/已预订/已取消）。
  - 管理用户账户，必要时可禁用违规账号。

- **系统设置**
  - 设置拼车费用和折扣规则。

- **问题处理**
  - 处理用户投诉和支持请求。

## **项目 2：拍卖平台（Auction Application）**

### **1. 系统概述**
本项目为**在线拍卖系统**，用户可以竞拍**古董和收藏品**，并提供专家认证和管理功能。

### **2. 主要功能**
#### **用户功能**
- **用户注册 & 认证**
  - 用户需要注册并登录。
  - 采用安全认证机制。

- **拍卖管理**
  - 用户可以上传拍卖品，设定起拍价和拍卖时长（最多 5 天）。
  - 用户可以参与竞拍，并跟踪竞价情况。
  - 当出价被超越或拍卖结束时，用户会收到通知。

- **物品认证**
  - 用户可以提交物品认证申请，提供图片和描述，由专家审核。

- **支付系统（模拟）**
  - 竞拍成功后进行模拟支付。
  - 用户可以查看交易记录。

#### **管理员功能**
- **拍卖收入管理**
  - 监控拍卖收入，设置佣金比例。

- **用户和拍卖管理**
  - 监督注册用户和拍卖活动。
  - 审核物品认证请求，分配专家审核。

- **专家管理**
  - 查看专家可用时间，并重新分配认证任务。

#### **专家功能**
- **物品审核**
  - 接收和处理认证请求。
  - 与用户沟通，要求补充资料。
  - 认证物品（批准、拒绝或请求二次审核）。

- **可用时间管理**
  - 设定未来 1-7 天的可用时间。
